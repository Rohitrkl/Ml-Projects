```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r"C:\\Users\\91840\\Downloads\\house-prices-advanced-regression-techniques\\train.csv")
```

```python
df.shape
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df.isnull().sum()
```

```python
null_var=df.isnull().sum()/df.shape[0]*100
null_var   
```

```python
drop_columns=null_var[null_var >20].keys()
drop_columns
```

```python
df2_drop_clm=df.drop(columns=drop_columns)
df2_drop_clm.shape
```

```python
df3_num=df2_drop_clm.select_dtypes(include=['int64','float64'])
df3_num.head()
    
```

```python
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()
```

```python
drop_columns= [var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
drop_columns
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(drop_columns):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':5, 'color':'#DC143C'})
```

```python
df4_num_mean=df3_num.fillna(df3_num.mean())
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(drop_columns):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df4_num_mean[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    plt.legend()
```
